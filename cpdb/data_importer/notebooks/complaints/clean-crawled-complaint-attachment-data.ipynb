```python
import requests
import json
import pandas
from tqdm import tqdm

from bs4 import BeautifulSoup

from data.models import AttachmentFile
from data_importer.ipra_portal_crawler.crawler import OpenIpraInvestigationCrawler, HEADERS, ComplaintCrawler
```

## Crawling complaint data from Ipra website
http://www.chicagocopa.org/wp-content/themes/copa/DynamicSearch.php

```python
links = OpenIpraInvestigationCrawler().crawl()
incidents = []
for link in tqdm(links):
    incidents.append(ComplaintCrawler(link).crawl())
```

## Clean up data

#### Read complaint attachments from crawled complaint data

```python
complaint_attachments = []
for incident in incidents:
    attachments = incident['attachments']
    crid = incident['log_number']
    
    for attachment in attachments:
        complaint_attachments.append({
            'crid': crid,
            'type': attachment['type'],
            'link': attachment['link'],
            'title': attachment['title']
        })

df = pandas.DataFrame.from_records(complaint_attachments)
df
```

#### Export data to csv file

```python
df.to_csv('ipra_crawled_complaint_attachments_data.csv', encoding='utf-8')
```
